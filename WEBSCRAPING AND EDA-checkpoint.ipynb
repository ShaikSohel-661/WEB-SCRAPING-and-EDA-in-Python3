```python
movie_name  = []
rating = []
metascore = []
gross = []
votes = []
year = []
duration = []
gener=[]
director=[]

Pagenum = []
for i in range(1,6):
    start_time = time.time()
    URL = 'https://www.imdb.com/list/ls050782187/?sort=user_rating,desc&st_dt=&mode=detail&page={}'.format(i)
    page = requests.get(URL)
    pagecontent = page.text
    soup = BeautifulSoup(pagecontent)
    
    for j in soup.find_all('div',attrs={'class':'lister-item-content'}):
        mv = j.find('h3',attrs = {'class':'lister-item-header'})
        movname=mv.find('a')
        ra = j.find('span',attrs = {'class':'ipl-rating-star__rating'})
        MS = j.find('div',attrs={'class':'inline-block ratings-metascore'})
        vt = j.find('span',attrs={'name':'nv'})
        yr = j.find('span',attrs={'class':'lister-item-year text-muted unbold'})
        gros = j.find('span',attrs = {'name':"nv"})
        gr = gros.find_next_sibling('span',attrs={'name':'nv'})
        dr = j.find('span',attrs={'class':'runtime'})
        gnr= j.find('span',attrs={'class':'genre'}).text
        x=j.find('p',attrs={'class':"text-muted text-small"})
        y=x.find_next_sibling('p',attrs={'class':"text-muted text-small"})
        drtr = y.find('a')
        
        if movname is None:
            movie_name.append(np.NaN)
        else:
            movie_name.append(movname.text.replace('\n',''))
            
        if ra is None:
            rating.append(np.NaN)
        else:
            rating.append(ra.text.replace('\n',''))
            
        if MS is None:
            metascore.append(np.NaN)
        else:
            metascore.append(MS.text.replace('\n','').replace('Metascore',""))
                             
        if gr is None:
            gross.append(np.NaN)
        else:
            gross.append(gr.text.replace('$','').replace('M',''))
            
        if yr is None:
            year.append(np.NaN)
        else:
            year.append(yr.text.replace('(','').replace(')',''))
            
        if dr is None:
            duration.append(np.NaN)
        else:
            duration.append(dr.text.replace('min',''))
            
        if gnr is None:
            gener.append(np.NaN)
        else:
            gener.append(gnr.replace('\n',''))
                        
        if vt is None:
            votes.append(np.NaN)
        else:
            votes.append(vt.text.replace(',',''))
            
        if drtr is None:
            director.append(np.NaN)
        else:
            director.append(drtr.text)
            
            
        Pagenum.append(i)
        
    end_time = time.time()
    print('page {} completed in {} seconds'.format(i,end_time-start_time))
    
    
```

```python
data = pd.DataFrame({'Title':movie_name,'Genre':gener,'Director':director,'Year':year,'Playtime (Minutes)':duration,'Rating':rating,'Votes':votes,'Gross (Millions)':gross,'Metascore':metascore})
```
